```python
# Define the Surfer Class
class Surfer():
    
    def __init__(self, name, hometown, rank):
        self.name = name
        self.hometown = hometown
        self.rank = rank
        
```

```python
# Create an instance of the Surfer Class

surfer = Surfer("Pepe", "Miami", "bottom")
surfer2 = Surfer("Paco", "Los Cabos", "professional")
surfer3 = Surfer("Mischa", "Ft Lauderdale", "star")

```

```python
# Print the object's attributes
print(surfer.name)
print(surfer.hometown)
print(surfer.rank)
```

```python
# ----BONUS----
# Variable to keep track of changes to while loop
go = True

# while loop runs so long as go is True
while go:
    
    name = input("Surfer's name? ")
    hometown = input("Surfer's hometown? ")
    rank = int(input("Surfer's rank? "))
    
    surfer = Surfer(name, hometown,rank)
    
    print(surfer.name)
    print(surfer.hometown)
    print(surfer.rank)
    
    check = input('Would u like to continue? (y/n) ')
    if (check.lower() == 'y'):
        go = True
    else:
        go = False


```
