```python
# Create bins in which to place values based upon TED Talk views
df = ted_df.groupby("views")

bins = [0, 9999999.9, 19999999.9, 29999999.9, 39999999.9, 49999999.9,]
# Create labels for these bins
group_names = ["F", "E", "D", "C", "B", "A"]
df.head()
```

```python
# Slice the data and place it into bins
df['Views Summary'] = pd.cut(df["views"], bins, labels=group_names, include_lowest=True)
df.head()
```
