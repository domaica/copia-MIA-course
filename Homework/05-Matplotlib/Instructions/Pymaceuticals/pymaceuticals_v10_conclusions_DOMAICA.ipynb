## Observations and Insights 

From analyzed data at least we can conclude the following remarks:

1.- The population by gender for this analysis is pretty homogeneus. Proportion of males and females samples is almost a perfect 50/50%.




2.- Sample sizes to conduct the analysis for total population looks to be well distributed and sufficient. Although Capomulin and Ramicane got a bigger sample, rest of drug samples are very homogeneous.




3.- According to boxplot analysis, we can observe that both Capomulin and Ramicane are almost equally effective reaching a much lower tumoral volume at the end of the 45 days treatment than Infubinol and Ceftamin. 

There are no outliers for this population. No observation falls more than above the upper quartile or below the lower quartile. This means that there are not any value in this population that shows abnormal distance from other values, therefore no observations need to be explained or removed from the mass of data. Data are consistent. 




4.- Capomulin drug seems to be very efective reducing volume tumors at end of 45 days period.




5.- Analyzing linear regression, we can observe a strong positive correlation (0.84) between mice weight and volume tumor size of 0.84. 
This means that the heavier the mice, the bigger is the tumor at end of ther treatment or that drug is less efective reducing tumor if the mouse weights more.

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties 
# of each drug regimen: 
summary_stats_df = clean_df.groupby('Drug Regimen')
# summary_stats_df.head()

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_vol_mean = summary_stats_df['Tumor Volume (mm3)'].mean()      # calculate average of tumor volume column
tumor_vol_median = summary_stats_df['Tumor Volume (mm3)'].median()  # calculate median of tumor volume column
tumor_vol_stdev = summary_stats_df['Tumor Volume (mm3)'].std()      # calculate standard deviation of tumor volume column
tumor_vol_sem = summary_stats_df['Tumor Volume (mm3)'].sem()        # calculate standard error of the mean (SEM)
                                                                    # of tumor volume column

# Assemble the resulting series into a single summary dataframe.
summary_groupby_df = pd.DataFrame({'Mean': tumor_vol_mean, 'Median': tumor_vol_median,
                                   'Std Deviation': tumor_vol_stdev, 'SEM': tumor_vol_sem})

summary_groupby_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
# THIS REQUEST IS KIND OF STRANGE AND PRODUCE STRANGE RESULTS, SEE BELOW EXPLANATION

# Aggregate every mean result for every drug and calculate mean of them
total_tumor_vol_mean = summary_groupby_df['Mean'].mean()
# Aggregate every median result for every drug and calculate median of this subset of results of previous table
total_tumor_vol_median = summary_groupby_df['Median'].median()

# in this result, I calculate the average of previous table standard deviations, because if I calculate the standard
# deviation of the standard deviations in the table, the result is 1.216, what is meaningless for std of population
# On the other hand, calculating average of STDs of table, result is 6.821, a more useful and meaningful value
total_tumor_vol_std = summary_groupby_df['Std Deviation'].mean()
# total_tumor_vol_std = summary_groupby_df['Std Deviation'].std()

# in this result, I calculate the average of previous table Standard Error of Means, because if I calculate the SEM
# of the SEMs subset in the table, the result is 0.033, what is meaningless for SEM of total population data
# On the other hand, calculating average of SEMs of table, result is 0.503, a more useful and meaningful value
total_tumor_vol_sem = summary_groupby_df['SEM'].mean()
# total_tumor_vol_sem = summary_groupby_df['SEM'].sem()

total_tumor_vol_mean = round(total_tumor_vol_mean, 3)
total_tumor_vol_median = round(total_tumor_vol_median, 3)
total_tumor_vol_std = round(total_tumor_vol_std, 3)
total_tumor_vol_sem = round(total_tumor_vol_sem, 3)

arr = np.array(['Mean:',total_tumor_vol_mean, '    ', 'Median:',total_tumor_vol_median, 
               '    ',  'Std Deviation:',total_tumor_vol_std,'    ', 'SEM:', total_tumor_vol_sem])
print(*arr, sep = ' ')
```

## Quartiles, Outliers and Boxplots

```python
# clean_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# generate dataframe from previous one keeping only columns needed for this calculations
reduced_df = clean_df[['Mouse ID','Drug Regimen','Timepoint', 'Tumor Volume (mm3)']]
# reduced_df
#-----------------------------------
# Select only certain drug regimens from previous dataframe and putting results in a new DF per
# 'Drug Regimen'

capomulin_df = reduced_df.loc[reduced_df["Drug Regimen"] == "Capomulin",:]
ramicane_df = reduced_df.loc[reduced_df["Drug Regimen"] == "Ramicane", :]
infubinol_df = reduced_df.loc[reduced_df["Drug Regimen"] == "Infubinol", :]
ceftamin_df = reduced_df.loc[reduced_df["Drug Regimen"] == "Ceftamin", :]

# append dataframes in a new dataframe
pick_drugs_df = capomulin_df
pick_drugs_df = pick_drugs_df.append(ramicane_df, ignore_index=True)
pick_drugs_df = pick_drugs_df.append(infubinol_df, ignore_index=True)
pick_drugs_df = pick_drugs_df.append(ceftamin_df, ignore_index=True)
  
# Start by getting the last (greatest) timepoint for each mouse
# Max it will be the max value for a Timepoint. Usually 45 days in this dataset
max_timepoint_df = pick_drugs_df[pick_drugs_df['Timepoint'] == pick_drugs_df['Timepoint'].max()]
# max_timepoint_df

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint--------------

# Rename columns to avoid conflicts
final_tumor_vol_df = max_timepoint_df.rename(columns={"Timepoint": "Max. timepoint", 
                                                               "Tumor Volume (mm3)": "Final tumor volume"})
# Drop unnecesary column that will be repeated after merging
final_tumor_vol_df = final_tumor_vol_df.drop(columns=['Drug Regimen'])
# final_tumor_vol_df

# Merge original dataframe with the df containing tumor volume at last timepoint
clean_data_merged_df = pd.merge(clean_df, final_tumor_vol_df, on='Mouse ID')

# Found in documentation that next option let show table in full. Can opt for .head() too
pd.set_option("display.max_rows", None, "display.max_columns", None)
clean_data_merged_df.head(20)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
list_tumor_vol = []

# Begin loop for drugs
for treatment in treatments:
    
    # new dataframe from previous one that contained volume at last timepoint per drug.
    # but getting only ones looping 
    drug_df = max_timepoint_df[max_timepoint_df['Drug Regimen'] == treatment]
      
    # Drop unnecesary column that will be repeated after merging. Timepoint is the max
    drug_df = drug_df.drop(columns=['Timepoint'])

    # Create a series that consists of all final volume values for each regimen
    final_vol = drug_df['Tumor Volume (mm3)']
    list_tumor_vol.append(final_vol)

    # Calculate the IQR and quantitatively determine if there are any potential outliers.

    quartiles = final_vol.quantile([.25,.5,.75]) # Set quartiles at 25%, 50%, 75%
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq        # Calculate interquartile range 3rd minus 1st quartile

    print(" ")     # empty line with space to separate drug results
    print(f'The interquartile range for {treatment} is: {iqr}')

    # Find upper and lower bounds to help identify outliers for each regimen ASTONISHING CONCLUSION
    lower_bound = lowerq - (1.5*iqr)           # lowerbound as defined in course to determine outliers below
    upper_bound = upperq + (1.5*iqr)           # upperbound as defined in course to determine outliers over
    print(f'The lower quartile for {treatment} is: {lowerq}')
    print(f'The upper quartile for {treatment} is: {upperq}')
    print(f"For drug {treatment} values below {lower_bound} could be outliers")
    print(f"For drug {treatment} values above {upper_bound} could be outliers")

    # Quantitative outliers. Use of PIPE to concatenate results for over upperbound and under lowerbound

    num_of_outliers = (final_vol.loc[(drug_df['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (drug_df['Tumor Volume (mm3)'] <= lower_bound)]).count()

    print(f'Number of outliers for {treatment} are: {num_of_outliers}')
       
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Boxplot
fig1, ax1 = plt.subplots(figsize=(10, 8))                # Boxplot and total size of graph
ax1.set_title('Tumor Volume by Treatment', fontsize =20) # Title and fontsize
ax1.set_ylabel('Final Tumor Volume (mm3)',fontsize = 16) # label for y axis and fontsize
ax1.set_xlabel('Drug Regimen',fontsize = 16)             # label for x axis and fontsize 
ax1.boxplot(list_tumor_vol, 
            widths = 0.4,
            patch_artist=True,  # fill with color
            vert=True)          # vertical box alignment

# text labels for ticks and fontsize
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'], fontsize = 14) 

plt.ylim(10, 80) # limit of y axis
# plt.grid()       
ax1.yaxis.grid(True) # Horizontal lines for reference 

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen


# Get weight and tumor volume from previous dataframe
weight = avg_vol_df['Weight (g)']
vol = avg_vol_df['Tumor Volume (mm3)']

# Calculate the correlation coefficient between mouse weight and average tumor volume
correlation = st.pearsonr(weight,vol)                 # pearson method
correlation = round(correlation[0],2)                 # round value with 2 decimals
# Print answer of above calculation
print(f"The correlation coeficient between weight and average tumor volume using Capomulin drug is {correlation}")

# Calculate linear regression for above values weight and tumor volume
(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, vol)
regress_values = weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2)) # equation of a line y = mx + b 

plt.figure(figsize=(14, 8))                                        # Set up total size of graph

# Plot linear regression on top of scatter plot ----------------------------------------------------
# Plot scatter
plt.scatter(weight,vol,
            s=100, 
            color="yellow",
            edgecolors='black')

# plot linear regression with variables
plt.plot(weight,regress_values,"r-")

# Show formula for linear regression
plt.annotate(line_eq,(18.1,41),              # position of line formula within x, y values in graph
             fontsize=14.5,
             color="red")  

# Add labels and title to plot
plt.xlabel("Weight (g)",fontsize =16)
plt.ylabel("Average Tumor Volume (mm3)",fontsize =16)
plt.title('Linear regression of mice weight vs average tumor volume using Capomulin',fontsize =18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid()


plt.show()
```
