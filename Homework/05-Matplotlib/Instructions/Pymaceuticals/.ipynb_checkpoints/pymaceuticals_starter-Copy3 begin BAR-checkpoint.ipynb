## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset keeping every value in all raws
mouse_dataset_df = pd.merge(mouse_metadata, study_results, how = "outer", on='Mouse ID' )


# Display the data table for preview
mouse_dataset_df
```

```python
# Checking the number of mice.
num_of_mice = mouse_dataset_df['Mouse ID'].nunique()
num_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_rows_df = mouse_dataset_df[mouse_dataset_df.duplicated(['Mouse ID', 'Timepoint'])]
# print("Duplicate Rows based on 2 columns are:", duplicate_rows_df, sep='\n')
print("Duplicate Rows based on 2 columns are:")
duplicate_rows_df
```

```python
# Optional: Get all the data for the duplicate mouse ID.
# duplicate_rows_df
duplicate_rows_df = mouse_dataset_df[mouse_dataset_df.duplicated(['Mouse ID', 'Timepoint'], 
                                                                 keep=False)]
duplicate_rows_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# result_df = source_df.drop_duplicates(keep=False)
# print('Result DataFrame:\n', result_df)

# clean_df = mouse_dataset_df.drop_duplicates(keep=False)
# clean_df

# mouse_dataset_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep=False)
# mouse_dataset_df

clean_df = mouse_dataset_df.loc[mouse_dataset_df["Mouse ID"] != "g989"]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
num_of_mice_clean_df = clean_df['Mouse ID'].nunique()
num_of_mice_clean_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties 
# of each drug regimen: 
summary_stats_df = clean_df.groupby('Drug Regimen')
# summary_stats_df.head()

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_vol_mean = summary_stats_df['Tumor Volume (mm3)'].mean()
tumor_vol_median = summary_stats_df['Tumor Volume (mm3)'].median()
tumor_vol_stdev = summary_stats_df['Tumor Volume (mm3)'].std()
tumor_vol_sem = summary_stats_df['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary dataframe.

summary_groupby_df = pd.DataFrame({'Mean': tumor_vol_mean, 'Median': tumor_vol_median,
                                   'Std Deviation': tumor_vol_stdev, 'SEM': tumor_vol_sem})

summary_groupby_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line

total_tumor_vol_mean = summary_groupby_df['Mean'].mean()
total_tumor_vol_median = summary_groupby_df['Median'].median()
total_tumor_vol_std = summary_groupby_df['Std Deviation'].mean()
total_tumor_vol_sem = summary_groupby_df['SEM'].mean()

total_tumor_vol_mean = round(total_tumor_vol_mean, 3)
total_tumor_vol_median = round(total_tumor_vol_median, 3)
total_tumor_vol_std = round(total_tumor_vol_std, 3)
total_tumor_vol_sem = round(total_tumor_vol_sem, 3)


# tumor_vol_mean = summary_stats_df['Tumor Volume (mm3)'].mean()
# tumor_vol_median = summary_stats_df['Tumor Volume (mm3)'].median()
# tumor_vol_stdev = summary_stats_df['Tumor Volume (mm3)'].std()
# tumor_vol_sem = summary_stats_df['Tumor Volume (mm3)'].sem()


# total_tumor_vol_mean
# total_tumor_vol_median
# total_tumor_vol_std
# total_tumor_vol_sem

arr = np.array(['Mean:',total_tumor_vol_mean, 'Median:',total_tumor_vol_median, 
                'Std Deviation:',total_tumor_vol_std,'SEM:', total_tumor_vol_sem])
print(*arr, sep = '   ')
```

```python
# Generate a bar plot showing the total number of measurements taken 
# on each drug regimen using pandas.
# ya teniamos groupedby y clean, lo reusamos
num_of_meas_by_drug = clean_df.groupby(["Drug Regimen"]).count()["Timepoint"] 
# num_of_meas_by_drug = pd.DataFrame(summary_stats_df.count())

plot_with_pandas = num_of_meas_by_drug.plot.bar(figsize=(10,6), color='r',
                                                 fontsize = 12, legend=False)
num_of_meas_by_drug
plt.xlabel("Drug Regimen",fontsize = 16, color='b')
plt.ylabel("Number of Treatments",fontsize = 16, color='b')
plt.title("Number of Measurements by Drug Regimen",fontsize = 18)
plt.tight_layout()
plt.show()
```
