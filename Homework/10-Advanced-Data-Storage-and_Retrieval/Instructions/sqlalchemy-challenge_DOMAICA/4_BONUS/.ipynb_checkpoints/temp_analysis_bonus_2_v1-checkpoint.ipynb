## Reflect Tables into SQLALchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session

# Addition inspect
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
# engine = create_engine("sqlite:///hawaii.sqlite")

engine = create_engine('sqlite:///../Resources/hawaii.sqlite')
```

```python
# reflect an existing database into a new model
# Declare a Base using `automap_base()`
Base = automap_base()

# reflect the tables
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## Bonus Challenge Assignment: Temperature Analysis II

```python
# Create function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
def calc_temps(start_date, end_date):
    
    # and return the minimum, maximum, and average temperatures for that range of dates 
    return session.query(func.min(measurement.tobs).label('min_temp'),\
                         func.max(measurement.tobs).label('max_temp'),\
                         func.avg(measurement.tobs.label('avg_temp'))).\
        filter(measurement.date >= start_date).filter(measurement.date <= end_date).all()

# Print results
print(calc_temps('2017-08-01', '2017-08-07'))
```

```python
# Use the function `calc_temps` to calculate the tmin, tavg, and tmax 

# Variables containing dates  for a year in the data set
start_date = '2016-01-01'
end_date = '2016-12-31'

# invoke previous function with input dates
year_temperatures = calc_temps(start_date, end_date)
# variables to print
min = year_temperatures[0][0]
max = year_temperatures[0][1]
avg = year_temperatures[0][2]
# round average obtained to 2 decimals
round_avg = str(round(avg, 2))

print(f"MINIMUM temperature btw {start_date} and {end_date} is {min} degrees F.")
print("")
print(f"MAXIMUM temperature btw {start_date} and {end_date} is {max} degrees F.")
print("")
print(f"AVERAGE temperature btw {start_date} and {end_date} is {round_avg} degrees F.")
```

### Daily Rainfall Average

#### Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.

```python
# Checking previous year variables are still there with correct values
start_date, end_date
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's 
# matching dates.

# Create dataframe from query of db gettind needed infos: date, station, precip for that dates
rain_df = pd.DataFrame(session.query(measurement.date, measurement.station, measurement.prcp).\
                               filter((measurement.date >= start_date) & \
                               (measurement.date <= end_date)).all())

rain_df
```

```python
# Calculation, of total precipitations sum values of precip grouping by station

rain_by_station_df = rain_df.groupby('station').sum()

rain_by_station_df
```

```python
# Now, we need to merge tables of measurements and stations
# as we do not have all data required. I need to bring the other stations table in this database to merge with this one
# check station table content and fields
first_row = session.query(station).first()
first_row.__dict__
```

```python
# list the station, name, latitude, longitude, and elevation
station_df = pd.DataFrame(session.query(station.station, station.name,\
                                        station.latitude, station.longitude,station.elevation).all())
station_df
```

```python
# merge stations table with precipitation sum by station on station
merged_df = pd.merge(rain_by_station_df, station_df, on="station")
merged_df
```

```python
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
sorted_df = merged_df.sort_values('prcp', ascending=False)
sorted_df
```

```python
# as you can see above, after sorting the dataframe, the index is wrong so we reset it to reorder.
sorted_df = sorted_df.reset_index(drop=True)
sorted_df
```

```python
# Use this function to calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)
# daily_normals("Month-Day") 'MM-DD'
#example 1st of august. initial day of our trip to Hawaii

# Variable with month and date
date_for_daily_normals = '08-01'

# function with input for month and date
def daily_normals(date_for_daily_normals):
    
    sel = [func.min(measurement.tobs), func.avg(measurement.tobs), func.max(measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", measurement.date) == date_for_daily_normals).all()
```

```python
# call function to be executed
daily_normals(date_for_daily_normals)
```

```python
# # Set the start and end date of the trip

start_date = '2017-08-01'
end_date = '2017-08-07'

# as usual, transform string to date format with datetime tools
# Y in capitals is year with 4 digits YYYY
# m in lowercase is month with 2 digits MM f.ex "01" as January
# d in lowercase is day expressed in 2 digits 01 to 31
start_date = datetime.strptime(start_date, '%Y-%m-%d')
end_date = datetime.strptime(end_date, '%Y-%m-%d')

# Use the start and end date to create a range of dates
# empty list for range of calculation dates
range_dates = []

# initialize 
date_of_record = start_date

# date or record is the date we are using to extract data
while date_of_record <= end_date:
    month = date_of_record.strftime('%m')
    day = date_of_record.strftime('%d')
    range_dates.append(f"{month}-{day}")
    date_of_record += dt.timedelta(days=1)

# Loop through the list of %m-%d strings and calculate the normals for each date
# empty list to append normals stats
normals = []

# loop with length of range of dates in this case a week or 7 days
for i in range(len(range_dates)):
    #invoke above function execution with range dates as input date
    normals_for_day = daily_normals(range_dates[i])
    # append info into list
    normals.append(normals_for_day[0])
    
# Load the previous query results into a Pandas DataFrame` Stats and dates
plot_df = pd.DataFrame(normals, columns =['tmin', 'tavg', 'tmax']) 
plot_df["date"] = range_dates

plot_df
```

```python
# Set `date' as index
plot_df = plot_df.set_index("date")
plot_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`

plot_df.plot.area(figsize=(20,10), stacked=False, fontsize=12)

# plt.xticks(np.arange(len(trip_dates)), labels=trip_dates)

plt.title(f"Daily Normals for Hawaii Trip", fontsize=30)
plt.xlabel("Dates", fontsize=25, color='b')
plt.ylabel("Temperature in degrees Farenheit", fontsize=25, color='b')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc='right', fontsize = 25)

# plt.tight_layout()
plt.savefig("images/bonus_2_normals_hawaii_vacation.png")
plt.show()
```
